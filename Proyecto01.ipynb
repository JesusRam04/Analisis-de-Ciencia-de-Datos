```python
import pandas as pd

# Supongamos que tienes un DataFrame llamado df

# Renombra las columnas
df = df.rename(columns={'Unnamed: 0':'product', 'Total':'Cadereyta total libreImpuestos', 'Unnamed: 2': 'Cadereyta Ctdad Ordenada', 'Unnamed: 3': 'Coahuila total libreImpuestos', 'Unnamed: 4': 'Coahuila Ctdad Ordenada',
       'Unnamed: 5': 'Durango total libreImpuestos', 'Unnamed: 6': 'Durango Ctdad Ordenada', 'Unnamed: 7':'Expos F total libreImpuestos', 'Unnamed: 8': 'Expos F Ctdad Ordenada', 'Unnamed: 9': 'Laredo total libreImpuestos',
       'Unnamed: 10': 'Laredo Ctdad Ordenada', 'Unnamed: 11': 'Monterrey total libreImpuestos', 'Unnamed: 12':'Monterrey Ctdad Ordenada', 'Unnamed: 13': 'Online total libreImpuestos',
       'Unnamed: 14': 'Online Ctdad Ordenada', 'Unnamed: 15': 'Queretaro total libreImpuestos', 'Unnamed: 16': 'Queretaro Ctdad Ordenada', 'Unnamed: 17': 'Reynosa total libreImpuestos',
       'Unnamed: 18': 'Reynosa Ctdad Ordenada', 'Unnamed: 19': 'San Luis total libreImpuestos', 'Unnamed: 20': 'San Luis Ctdad Ordenada', 'Unnamed: 21':'Total libreImpuestos',
       'Unnamed: 22': 'Ctdad Ordenada'})

# Muestra el DataFrame con las columnas renombradas
print(df)
```

```python
def cambiar_formato_fecha(fecha):
    partes = fecha.split()  # Dividir la fecha en partes: ['01', 'ene', '2024']
    if partes[1] == 'abr':
        nuevo_formato = partes[0] + '-' + '04' + '-' + partes[2]  # Construir el nuevo formato: '01-abr-2024'
    elif partes[1] == 'may':
        nuevo_formato = partes[0] + '-' + '05' + '-' + partes[2]  # Construir el nuevo formato: '01-may-2024'
    elif partes[1] == 'jun':
        nuevo_formato = partes[0] + '-' + '06' + '-' + partes[2]  # Construir el nuevo formato: '01-jun-2024'
    return nuevo_formato

# Aplicar la función para cambiar el formato de las fechas a la columna 'fecha'
df['fecha'] = df['fecha'].apply(cambiar_formato_fecha)
```

```python
df['fecha'] = pd.to_datetime(df['fecha'], format='%d-%m-%Y')
```
