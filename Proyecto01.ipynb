```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_excel('vtas abr-Jun22.xls')
df
```

```python
df = df.drop(df.index[:3])
```

```python
df['fecha'] = np.nan
```

```python
df
```

```python
import re
pattern = r'\s*\d{2}\s+(ene|feb|mar|abr|may|jun|jul|ago|sep|oct|nov|dic)\s+\d{4}\s*'
for i in range(len(df)):
    if re.match(pattern, str(df.iloc[i,0])):
        fecha = df.iloc[i,0]
    else:
        df.iloc[i,-1] = fecha
```

```python
df
```

```python
df = df.dropna(subset=['fecha'])
```

```python
df.fecha.unique()
```

```python
df['fecha'] = df['fecha'].str.strip()
df.fecha.unique()
```

```python
df['fecha'] = pd.to_datetime(df['fecha'], format='%d %b %Y').dt.strftime('%d-%m-%Y')
```

```python
meses = {
    'ene': 'enero',
    'feb': 'febrero',
    'mar': 'marzo',
    'abr': 'abril',
    'may': 'mayo',
    'jun': 'junio',
    'jul': 'julio',
    'ago': 'agosto',
    'sep': 'septiembre',
    'oct': 'octubre',
    'nov': 'noviembre',
    'dic': 'diciembre'
}

# Crear un DataFrame de ejemplo
# Convertir el formato de mes abreviado al formato completo en español
# Convertir el formato de mes abreviado al formato completo en español
df['fecha'] = df['fecha'].str.replace(r'\b(\s*\d{2}\s+\w{3}\s+\d{4})\b', lambda x: x.group(1).strip(), regex=True)
df['fecha'] = df['fecha'].str.replace(r'(\b\w{3}\b)', lambda x: meses.get(x.group(1)), regex=True)

# Convertir la columna 'Fecha' al tipo datetime
df['fecha'] = pd.to_datetime(df['fecha'], format='%d %B %Y', errors='coerce')
```
