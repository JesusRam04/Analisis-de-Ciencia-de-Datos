```python
def cambiar_formato_fecha(fecha):
    partes = fecha.split()  # Dividir la fecha en partes: ['01', 'ene', '2024']
    meses = {
        'ene': '01', 'feb': '02', 'mar': '03', 'abr': '04',
        'may': '05', 'jun': '06', 'jul': '07', 'ago': '08',
        'sep': '09', 'oct': '10', 'nov': '11', 'dic': '12'
    }
    mes_numero = meses.get(partes[1].lower())  # Obtener el número de mes del diccionario
    nuevo_formato = partes[0] + '-' + mes_numero + '-' + partes[2]  # Construir el nuevo formato: '01-abr-2024'
    return nuevo_formato

# Aplicar la función para cambiar el formato de las fechas a la columna 'fecha'
df['date'] = df['date'].apply(cambiar_formato_fecha)
```

```python
# Calcular la mediana de la serie temporal
median_value = product12.median()
# Definir un umbral para identificar los valores atípicos
threshold = 3.0  # Este es un valor arbitrario, puedes ajustarlo según tus datos y necesidades
# Calcular la desviación estándar de la serie temporal
std_dev = product12.std()
# Crear una máscara booleana para identificar los valores atípicos
outliers_mask = (product12 - median_value).abs() > threshold * std_dev
# Reemplazar los valores atípicos por la mediana de la serie temporal
product12[outliers_mask] = median_value
```

```python
from statsmodels.tsa.holtwinters import SimpleExpSmoothing, Holt
smooth = product12.copy()
smooth = SimpleExpSmoothing(np.asarray(product12))

smooth._index = pd.to_datetime(product12.index)
smooth = smooth.fit(smoothing_level=0.1).fittedvalues

smoothed_time_series = pd.Series(smooth, index=product12.index)

# Gráfico de la serie original y suavizada
plt.figure(figsize=(24, 6))
plt.plot(product12, label='Original')
plt.plot(smoothed_time_series, label='Smoothed')
plt.title('Original vs. Smoothed Time Series')
plt.legend()
plt.show()
```

```python
import pandas as pd
from statsmodels.tsa.seasonal import seasonal_decompose

decompose_data = seasonal_decompose(smoothed_time_series, model="additive")
decompose_data.plot()
```

```python
from statsmodels.tsa.stattools import adfuller
dftest = adfuller(smoothed_time_series, autolag = 'AIC')
print("1. ADF : ",dftest[0])
print("2. P-Value : ", dftest[1])
print("3. Num Of Lags : ", dftest[2])
print("4. Num Of Observations Used For ADF Regression and Critical Values Calculation :", dftest[3])
print("5. Critical Values :")
for key, val in dftest[4].items():
    print("\t",key, ": ", val)
```

```python
from statsmodels.tsa.stattools import acf, pacf
from statsmodels.graphics.tsaplots import plot_acf, plot_pacf

lag_acf = acf(smoothed_time_series, nlags=20)
lag_pacf = pacf(smoothed_time_series, nlags=20, method='ols')

fig, ax = plt.subplots(1,2,figsize=(16,4))
plot_acf(lag_acf, ax=ax[0])
plot_pacf(lag_pacf, ax=ax[1])


plt.show()
```

```python
day_of_week = smoothed_time_series.index.weekday

# Convierte la serie de días de la semana en una variable exógena DataFrame
exog_variable_weekday = pd.DataFrame(day_of_week, index=smoothed_time_series.index)
```

```python
# Obtén el número del mes para cada fecha en la serie temporal
month_number = smoothed_time_series.index.month

# Convierte la serie de números de mes en una variable exógena DataFrame
exog_variable_month = pd.DataFrame(month_number, index=smoothed_time_series.index)

# Ahora 'exog_variable_month' contiene el número del mes como una variable exógena
```

```python
exog_variables = pd.concat([exog_variable_month, exog_variable_weekday], axis=1)
```

```python
from sklearn.model_selection import train_test_split

train_exog, test_exog = train_test_split(exog_variables, test_size = 0.10, shuffle = False)
```

```python
import statsmodels.api as sm

model=sm.tsa.statespace.SARIMAX(train, exog=train_exog, order=(1,0,0),seasonal_order=(1,0,1,7))
results=model.fit()
predictions = results.predict(start='2023-10-19', end='2023-12-30', exog = test_exog)
```

```python
print(results.summary())
```

```python
from sklearn.metrics import mean_squared_error

error = mean_squared_error(test, predictions)
print('Test MSE: %.5f' % error)
```

```python
from sklearn.metrics import r2_score 

error = r2_score(test, predictions)
print('Test COD: %.5f' % error)
```

```python
import numpy as np

n = len(test)
residuals = test - predictions
rse = np.sqrt(np.sum(residuals**2) / (n - 2))
print("RSE:", rse)
```
